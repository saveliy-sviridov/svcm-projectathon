Recherche d'un lot de soumission (`List`) pour le patient `279035121518980` (système `urn:oid:1.2.250.1.213.1.4.10`), daté du 14 janvier 2013, statut `current`.

## Configuration

```python
import os
from datetime import datetime
from urllib.parse import quote

NOTEBOOK_ID = "07_pdsm_flux5a_n1"
RUN_TS  = datetime.now().strftime("%Y%m%dT%H%M%S")
RUN_DIR = os.path.join("runs", f"{NOTEBOOK_ID}_{RUN_TS}")
os.makedirs(RUN_DIR, exist_ok=True)

print(f"Configuration OK — sortie dans : {RUN_DIR}")
```

---
## Étape 0 — Construction et exécution de la requête

**Requête** : `GET /fhir/List?code=...submissionset&patient.identifier=...&date=2013-01-14&status=current`

```python
PATIENT_SYSTEM    = "urn:oid:1.2.250.1.213.1.4.10"
PATIENT_ID        = "279035121518980"
SUBMISSION_DATE   = "2013-01-14"
STATUS            = "current"
MHD_LIST_CODE_SYS = "https://profiles.ihe.net/ITI/MHD/CodeSystem/MHDlistTypes"
MHD_LIST_CODE     = "submissionset"

url_search = (
    f"{{pdsm-server}}/List"
    f"?code={quote(MHD_LIST_CODE_SYS)}|{MHD_LIST_CODE}"
    f"&patient.identifier={quote(PATIENT_SYSTEM)}|{PATIENT_ID}"
    f"&date={SUBMISSION_DATE}"
    f"&status={STATUS}"
    f"&_format=json"
)

print("[Étape 0] Requête construite :")
print(f"\n  {url_search}\n")
```
